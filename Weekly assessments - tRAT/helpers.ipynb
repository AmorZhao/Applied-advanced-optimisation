```python
import matplotlib.pyplot as plt
import numpy as np

# ----------------- Plotting -----------------

def plotHistograme(data):
    plt.hist(data, bins=50)
    plt.show()

def plotHistogrameWithMultipleDataset(
    data, 
    label = None, 
    legend = None
):
    for dataset in data:
        plt.hist(dataset, bins=50)

    if label:
        plt.plot(label = label)
    if legend:
        plt.legend(legend)

    plt.show()

def plotWithMask(
    x_data, 
    y_data,
    mask, 
    label = None, 
    legend = None
):
    plt.plot(x_data[~mask], np.array(y_data)[~mask], 'r.')
    plt.plot(y_data[mask], np.array(y_data)[mask], 'b.')

    if label:
        plt.plot(label=label)
    if legend:
        plt.legend(legend)
    
    plt.show()


# ----------------- CVPXY -----------------

import cvxpy as cp 

def solveOptimizationProblem(
    constraints,
    objective, 
    isMaximization = False
):
    problem  = cp.Problem(cp.Minimize(objective), constraints)
    if isMaximization:
        problem  = cp.Problem(cp.Maximize(objective), constraints)
    problem.solve()
    
    print("optimal value: ", problem.value)

```
