Consider the basic norm approximation problem

$$\displaystyle\min_x || Ax-b || \tag{1}$$

Tha data of the problem is given by

```python
import cvxpy as cp
import numpy as np

m = 100
n = 30
np.random.seed(1)
A = np.random.randn(m, n)
b = np.random.randn(m)
```

**Question 1:** Solve problem $(1)$ for the norms



1.   $\ell_1$-norm
2.   $\ell_2$-norm
3.   $\ell_\infty$-norm

You may find this useful https://www.cvxpy.org/tutorial/functions/index.html#scalar-functions

```python
x1 = cp.Variable(n)
x2 = cp.Variable(n)
x3 = cp.Variable(n)

cost_1 = cp.norm(A @ x1 - b, p=1)
constraints = []

prob_1 = cp.Problem(cp.Minimize(cost_1), constraints) 
prob_1.solve() 
print("optimal value", prob_1.value)
print("optimal var", x1.value)


cost_2 = cp.norm(A @ x2 - b, p=2)
prob = cp.Problem(cp.Minimize(cost_2), constraints) 
prob.solve() 
print("optimal value", prob.value)
print("optimal var", x2.value)


cost_3 = cp.norm(A @ x3 - b, "inf")
prob = cp.Problem(cp.Minimize(cost_3), constraints) 
prob.solve() 
print("optimal value", prob.value)
print("optimal var", x3.value)
```

**Question 2:** Compute the residuals for each of the three norms (e.g. r_l1 = A @ x.value - b) and plot the three hystograms of the residuals. To plot the hystograms use the command

```python
residue_1 = A @ x1.value - b
residue_2 = A @ x2.value - b
residue_3 = A @ x3.value - b

print("residue_1", np.linalg.norm(residue_1, 1))
print("residue_2", np.linalg.norm(residue_2, 2))
print("residue_3", np.linalg.norm(residue_3, np.inf))
```

```python
import matplotlib.pyplot as plt

plt.hist(residue_1, 50)
# plt.show()
plt.hist(residue_2, 50)
# plt.show()
plt.hist(residue_3, 50)
plt.show()
```
